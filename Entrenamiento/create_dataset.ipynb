```python
import pandas as pd
```

```python
df = pd.read_csv("data/labels.csv")
df.head()
```

```python
df_labels = df.drop(["AUDIO","Etiquetador"], axis=1)
df_labels.head()
```

```python
df_text = pd.read_csv("data/output_text.txt", header=None, delimiter=":")
df_text.head()
```

```python
df_full = df_labels
df_full["Text"] = df_text[1]
columns = ['Text','Producto', 'Intención', 'Tipo de movimiento',
           'Contexto 1', 'Contexto 2', 'Detalle 1', 'Verbalizacion del cliente']
df_full = df_full.reindex(columns,axis=1)
df_full
```

```python
df_full = df_full.fillna("Otro")
df_full.head()
```

```python
df_full.describe()
```

```python
df_full.to_csv("data/dataset.csv", index=False)
```
